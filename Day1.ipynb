```python
#import
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.arange(10,21)
b = np.arange(20,31)

plt.title("Question 1:", color='Red')
plt.xlabel("X-Axis", color='Blue')
plt.ylabel("Y-Axis", color='Brown')
plt.plot(a,b,'g*-',linewidth=1,markersize=14)
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
sales_1 = [140,150,140,145,15,65,200] #company1
sales_2 = [170,90,60,50,240,45,195] #company2

plt.title("Question 2: ",color='blue')
plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')

plt.plot(days, sales_1, 'r*-', markersize=12, linewidth=2)
plt.plot(days, sales_2, 'go-', markersize=12, linewidth=2)
plt.show()
```
